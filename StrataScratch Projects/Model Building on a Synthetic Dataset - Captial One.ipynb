```python
cols_with_nulls = df.columns[df.isnull().any()]
prepared_df = df.copy()
prepared_df[cols_with_nulls] = prepared_df[cols_with_nulls].fillna(prepared_df[cols_with_nulls]\
                                                           .mean(numeric_only=True))
```

```python
nullvalue_columns = prepared_df.isnull().sum()\
                               .sort_values(ascending=False)\
                               .head(20)
nullvalue_columns.head()
```

```python
for col in prepared_df.select_dtypes(include='object'):
    prepared_df[col] = prepared_df[col].fillna(prepared_df[col].mode()[0])

prepared_df.isnull().sum()
```
